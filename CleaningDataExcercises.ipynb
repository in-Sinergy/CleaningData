# **EXCERCISE 1: CARBON EMISSIONS**

```python
# importamos la información desde el enlace
import pandas as pd

linkToFile='https://github.com/CienciaDeDatosEspacial/code_and_data/raw/main/data/carbonEmi_downloaded.csv'
carbon=pd.read_csv(linkToFile)
```

```python
#leemos la database
carbon
```

```python
#revisamos el tipo de información que nos proporciona la tabla
carbon.info()
```

```python
#creamos tablas de frecuencia previas para poder identificar donde y qué limpiar
carbon.region.value_counts()
```

```python
carbon.date_of_information.value_counts()
```

## 1. Keep the columns name, value, date_of_information and region.
Tip: use drop, loc, and iloc for the same purpose (three ways to accomplish the task).

```python
#debemos borrar las columnas de slug y ranking, así que creamos una variable que las contenga
deleteColumn=['slug', 'ranking']
carbon.drop(columns=deleteColumn, inplace=True)
carbon
```

## 2. Change the column name date_of_information to carbon_date.
Tip: Use rename.

```python
from os import rename
#cambiamos el nombre de la columna
copy =carbon.copy()
copy.rename(columns={'date_of_information' : 'carbon_date'}, inplace=True)
copy
```

```python
#aplicamos lo logrado al database original
carbon.rename(columns={'date_of_information' : 'carbon_date'}, inplace=True)
carbon
```

## 3. Make sure the cells with text does not have neither trailing nor leading spaces.
Tip: use strip.

```python
#las celdas con valores de texto son 'name' y 'region'
carbon.name.str.strip()
```

```python
carbon.region.str.strip()
```

```python
carbon['name']=carbon.name.str.strip()
```

```python
carbon['region']=carbon.region.str.strip()
```

```python
carbon
```

##4. Detect the presence of symbols in the numeric data that are not numeric or point.
Tip: Use contains.

```python
#columnas de data numérica: value, carbon_date
carbon.value.str.contains(pat=r'\D',
                          regex=True).value_counts()
```

```python
carbon.carbon_date.str.contains(pat=r'\D',
                          regex=True).value_counts()
```

```python
#de acuerdo con los resultados obtenidos, todas las celdas continen elementos no numéricos en las columnas analizadas
```

##5. Make sure there are no spaces as part of the column names.
Tip: use replace.

```python
carbon.columns.to_list()
#como podemos ver, las columnas no tienen espacios en el nombre
```

## 6. Get rid of any value detected in the previous step:
Tip: use replace.


```python
#debemos corregir la columna value y eliminar los símbolos no numéricos
carbon.value.str.replace(pat=r',', repl='', regex=True)
```

```python
carbon.value.str.extract(pat=r'(\d+)',expand=True)
```

```python
#observamos el resultado
carbon['value']=carbon.value.str.replace(pat=r',', repl='', regex=True).\
                              str.extract(pat=r'(\d+)',expand=True)
carbon
```

## 7. Keep only the year value in the column carbon_date.
Tip: use extract.

```python
#nos quedamos únicamente con el valor numérico de la columna carbon_date
carbon.carbon_date.str.extract(pat=r'(\d+)',expand=True)
```

```python
#lo aplicamos al database original y observamos el resultado
carbon['carbon_date']= carbon.carbon_date.str.extract(pat=r'(\d+)',expand=True)
carbon
```

## Resultados finales

```python
#verificamos la presencia de elementos no numéricos nuevamente:
carbon.value.str.contains(pat=r'\D',
                          regex=True).value_counts()
```

```python
carbon.carbon_date.str.contains(pat=r'\D', regex=True).value_counts()
```

# **EXCERCISE 2:  Revenue from forest resources**

```python
from IPython.display import IFrame
ciaLink2="https://www.cia.gov/the-world-factbook/field/revenue-from-forest-resources/country-comparison"
IFrame(ciaLink2, width=900, height=900)
```

```python
# read web table into pandas DF
import pandas as pd

forestDFs=pd.read_html(ciaLink2, # link
                        header=0, # where is the header?
                        flavor='bs4')
```

```python
# make a copy
forest=forestDFs[0].copy()
```

```python
forest
```

```python
forest.columns.to_list()
```

```python
forest['Date of Information'].value_counts()
```

## 1. Replace '%' by 'pct'.
Tip: use replace.

```python
from os import rename
#cambiamos el nombre de la columna
forest.rename(columns={'% of GDP' : 'pct of GDP'}, inplace=True)
forest
```

## 2. Keep the columns Country, pct of GDP, and Date of Information.
Tip: use drop, loc, and iloc for the same purpose (three ways to accomplish the task).

```python
#debemos borrar la columna rank, en este caso usando la función drop
deleteColumn=['Rank']
forest.drop(columns=deleteColumn, inplace=True)
forest
```

## 3. Change the column name Date of Information to forest_date.
Tip: Use rename.

```python
#cambiamos el nombre de la columna
forest.rename(columns={'Date of Information' : 'forest_date'}, inplace=True)
forest
```

## 4. Make sure there are no spaces as part of the column names.
Tip: use replace.

```python
#revisamos la presencia de espacios en los índices de la tabla
forest.columns.str.contains(pat=r'\s', regex=True)
```

```python
#debemos corregir lo que ocurre en la segunda columna (pct of GDP)
forest.columns.str.replace(pat=r'\s', repl='_',regex=True)
```

```python
#aplicamos la nueva lista de índices
forest.columns = forest.columns.str.replace(pat=r'\s', repl='_',regex=True)
forest
```

## 5. Make sure the cells with text does not have neither trailing nor leading spaces.
Tip: use strip.

```python
#revisamos la columna country, que es la única con data de texto
forest.Country.str.strip()
```

```python
#confirmamos que no quedan espacios antes o después del texto
forest.Country.str.contains(pat=r'^\s | \s$', regex=True).value_counts()
```

## 6. Keep only the year value in the column forest_date.
Tip: use extract.

```python
forest.forest_date.str.extract(pat=r'(\d+)',expand=True)
```

```python
#lo aplicamos al database original
forest['forest_date']= forest.forest_date.str.extract(pat=r'(\d+)',expand=True)
forest
```

```python
#verificamos que no queden caracteres no numéricos
forest.forest_date.str.contains(pat=r'\D', regex=True).value_counts()
```
